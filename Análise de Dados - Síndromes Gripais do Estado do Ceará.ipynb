## Removendo espaços em brancos do nome dos municípios, colocando em caixa baixa e criando coluna com os novos nomes.

```python
municipios = dados.municipio
municipios
    
```

```python
nome_municipios=[]
for nome in municipios:
    nome=nome.lower().replace(' ','_')
    nome_municipios.append(nome)
    
```

```python
len(nome_municipios)
```

```python
dados['nome_municipio'] = nome_municipios
```

## Ao fazer o 'groupby' percebi dados de datas 'outliers', então resolvi agrupar por Ano e Mês

```python
dados['ano'] = dados['dataNotificacao'].dt.year 
dados['mes'] = dados['dataNotificacao'].dt.month 
```

## Fixando os dados apenas de 2022

```python
data_2022 = dados.loc[(dados['ano']) > 2021]
data_2022
```

```python
janeiro = data_2022.loc[((data_2022['mes']) == 1.0) & ((data_2022['totalTestesRealizados']) >= 0) ]
janeiro
```

## Percebemos que foram notificadas 342218 pessoas com síndromes gripais em janeiro

## Agrupando por cidades

```python
notificacao_por_mun = janeiro.groupby(['nome_municipio'], as_index=False).sum()
notificacao_por_mun
```

```python
soma = notificacao_por_mun['mes'].sum()
print(soma)
```

```python
notificacao_por_mun['porcentagem']= ((notificacao_por_mun['mes']/342218) * 100).round(2)
notificacao_por_mun.head()
```

```python
mun_mais_notificados = notificacao_por_mun.loc[((notificacao_por_mun['porcentagem']) > 2)]
#ordenando os dados em ordem decrescente
mun_mais_notificados = mun_mais_notificados.sort_values(by=['porcentagem'], ascending=False)
mun_mais_notificados
```

```python
outros = (100 - mun_mais_notificados.porcentagem.sum()).round(2)
outros
```
